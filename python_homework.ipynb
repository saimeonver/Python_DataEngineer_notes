````python
## 7.
#Add the odd and even numbers in the list `[22, 36, 45, 62, 39, 10, 25]` into two different lists.
````

````python
liste = [22, 36, 45, 62, 39, 10, 25]
````

````python
odd_list=[]
even_list=[]
````

````python
for i in liste:
  if i % 2 == 0:
    even_list.append(i)
  else:
    odd_list.append(i)

print(f"Even list: {even_list}")
print(f"Odd list: {odd_list}")
````
