Buradan itibaren yazılınca metin sırayla ekleniyor


```python
with open("new_file.txt","w") as f:
  f.write("First line\n")
```

```python
with open("new_file.txt","a") as f:
  f.write("Second line\n")
```

```python
with open("new_file.txt","r") as f:
  print(f.read())
```
